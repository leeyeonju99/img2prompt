```python
gen_data_path = Path("/kaggle/input/generatedimages/dataset")
```

```python
def image_id2path_new_dataset(img_id: str) -> str:
    return str(gen_data_path / "images" / f"{img_id}.png")
```

```python
sys.path.append('../input/sentence-transformers-222/sentence-transformers')
from sentence_transformers import SentenceTransformer, models
```

```python
st_model = SentenceTransformer('/kaggle/input/sentence-transformers-222/all-MiniLM-L6-v2')
```

```python
prompt_embeddings = st_model.encode(prompts['prompt'])
```

```python
prompt_embeddings.shape
```

```python
similarity_matrix = np.inner(prompt_embeddings, prompt_embeddings)
```

```python
pd.DataFrame(similarity_matrix)
```

```python
threshold = 0.95
similar_prompts = np.where(similarity_matrix > threshold)
unique_prompts = np.unique(similar_prompts[0])
df_filtered = prompts.iloc[unique_prompts]
```

```python
indices = np.where(similarity_matrix > 0.95)
unique_indices = set(np.unique(indices))
prompts_filtered = prompts.drop(list(unique_indices))

print(prompts_filtered)
```
